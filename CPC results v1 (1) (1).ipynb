```python
df2=df.drop(columns=['Nom de la circonscription', 'Type de résultats**', 'Middle name(s) - Autre(s) prénom(s)', 'Appartenance politique'])
df3 = df2.drop([df2.index[4035],df2.index[4036],df2.index[4037]])
df4 = df3[df3['Type of results*'] != 'preliminary']
df5 = df4.replace({'Conservative':'CPC','NDP-New Democratic Party':'NDP','Liberal':'LPC',"People's Party - PPC":'PPC'},regex=True)
df5.reset_index(inplace = True)
df5.drop(df5.loc[df5['Electoral district name']=='Dartmouth--Cole Harbour'].index, inplace=True)
df5
#with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    #print(df5)
```

```python
dict_ridings = dict()
for k, v in df5.groupby('Electoral district name'):
    dict_ridings[k] = v
    
dict_ridings['Bay of Quinte']
```

```python
#for key, value in dict_ridings.items() :
#    print (key)
#len(dict_ridings)
#removed_value = dict_ridings.pop('Dartmouth--Cole Harbour')
#dict_ridings
```

```python
ridings=[]
margins=[]
close_losses_ridings={}
margin = 10

for key in dict_ridings:
    riding_df=dict_ridings[key]
    percent_of_votes_won = riding_df['% Votes obtained - Votes obtenus %']
    winning_percent = percent_of_votes_won.max()
    cpc_vote_percent = riding_df.loc[riding_df['Political affiliation']=='CPC']['% Votes obtained - Votes obtenus %'].iloc[0]
    difference = winning_percent - cpc_vote_percent
    #print(key,difference)
    #difference_int = int(difference)
    ridings.append(key)
    margins.append(difference)
    #print('riding is:', key,'difference is:', difference)
    if difference > 0 and difference < 10:
        close_losses_ridings[key] = difference
    #else:
        #print('-')
        
close_losses_ridings
```

```python
close_victories_ridings={}

for key in dict_ridings:
    riding_df=dict_ridings[key]
    percent_of_votes_won = riding_df['% Votes obtained - Votes obtenus %']
    winning_percent = percent_of_votes_won.max()
    cpc_vote_percent = riding_df.loc[riding_df['Political affiliation']=='CPC']['% Votes obtained - Votes obtenus %'].iloc[0]
    lpc_vote_percent = riding_df.loc[riding_df['Political affiliation']=='LPC']['% Votes obtained - Votes obtenus %'].iloc[0]
    ndp_vote_percent = riding_df.loc[riding_df['Political affiliation']=='NDP']['% Votes obtained - Votes obtenus %'].iloc[0]
    difference = winning_percent - cpc_vote_percent
    lpc_difference = cpc_vote_percent - lpc_vote_percent
    ndp_difference = cpc_vote_percent - ndp_vote_percent
    #print(key,difference)
    #difference_int = int(difference)
    #ridings.append(key)
    #margins.append(difference)
    #print('riding is:', key,'difference is:', difference)
    if difference == 0 and (lpc_difference < 10 or ndp_difference < 10):
        close_victories_ridings[key] = min(lpc_difference,ndp_difference)

close_victories_ridings
```

```python
(pd.DataFrame.from_dict(data=close_losses_ridings, orient='index')
   .to_csv(r'C:\Users\hknifton\Downloads\close_losses.csv', header=False))
```

```python
(pd.DataFrame.from_dict(data=close_victories_ridings, orient='index')
   .to_csv(r'C:\Users\hknifton\Downloads\close_wins.csv', header=False))
```
